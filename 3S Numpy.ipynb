#### 27. Рассмотрим случайный вектор формы (100,2), представляющий координаты, найдите расстояния между точками. (★★☆)

````python
def euqDistPtToArr(pt1Arr, pt2):
    #sqrt((x1 - x2)**2 + (y1 - y2)**2) 
    return np.sqrt(np.square(pt1Arr[:,0].T - pt2[0]) + np.square(pt1Arr[:,1].T - pt2[1]))
````

````python
mx100x2 = np.random.random((100, 2))
nSize = mx100x2.shape[0]
mxDist = np.ones((nSize, nSize))

for idx in np.arange(nSize):
    mxDist[idx,:] = euqDistPtToArr(mx100x2, mx100x2[idx])

with np.printoptions(threshold=np.inf):
    np.disp(mxDist)
````

#### 38. Рассмотрим 2 набора точек P0,P1, описывающих линии (2d), и набор точек P. Как вычислить расстояние от каждой точки j (P[j]) до каждой линии i (P0[i],P1[i])?(★★★)

````python
def x(pt):
    return pt[0]
    
def y(pt):
    return pt[1]

def solve(pt1, pt2):
    a = np.row_stack((x(pt1), x(pt2)))
    a = np.column_stack((a, np.ones_like(pt1)))
    b = np.row_stack((y(pt1), y(pt2)))
    return np.linalg.solve(a, b)

def logBool(log, txt, obj):
    if log:
        print(txt, obj)

def cross(pt1, pt2, pt3, log=False):
    line = solve(pt1, pt2)
    if log:
        logBool(log, "Line i from dot P0[i] P1[i]:", line)

    t = y(line) / x(line)
    perp = np.array([t, -t * x(pt3) + y(pt3)])
    if log:
        logBool(log, "Perpendicular to line:", perp)

    m = np.array([[-1], [1]])
    pCrossed = solve(line * m, perp * m)
    if log:
        logBool(log, "Crossed line w perp:", pCrossed)

    maxP, minP = np.max([x(pt1), x(pt2)]), np.min([x(pt1), x(pt2)])
    if x(pCrossed) >= minP and x(pCrossed) <= maxP:
        return pCrossed
    return np.array([np.inf, np.inf]) # retunr inf if out of line
````

````python
nL, nP = 3, 10
P0, P1, P = np.random.random((nL, 2)), np.random.random((nL, 2)), np.random.random((nP, 2))

mx100x2 = np.random.random((nL, 2))
mxDist = np.ones((nP, nL))

for idx in np.arange(nP):
    #dist to one side
    distP0 = euqDistPtToArr(P0, P[idx])
    
    # dist to other side
    distP1 = euqDistPtToArr(P1, P[idx])
    
    # dist perpendicular to between if exist 
    distCross = np.zeros((nL,2), dtype=np.float64)
    for idxP in np.arange(nL):
        distCross[idxP,:] = cross(P0[idxP], P1[idxP], P[idx], False).ravel()
    distCrossP = euqDistPtToArr(distCross, P[idx])
    
    # set minimum to result
    mxDist[idx,:] = np.array([distP0, distP1, distCrossP]).min(axis=0)

    print("Line ID:", idx, "Dist to P0:", distP0)
    print("Line ID:", idx, "Dist to P1:", distP1)
    print("Line ID:", idx, "Dist to PCross:", distCrossP)
    print("Line ID:", idx, "MIN Dist:", mxDist[idx,:])

print("\nResult (Min distanse from each P to each line):")
with np.printoptions(threshold=np.inf):
    np.disp(mxDist)
````
